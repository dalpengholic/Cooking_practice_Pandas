## 2. Brief checking of a dataframe

When you read a csv, you might want to see how it looks like. Here, some good methods to be used. When you make a dataframe instance, you can use methods and attributes using `dataframe name.` like `df_1.head()`, `df_1["column name"]`

```python
# Use comment and uncomment to execute each line
df_1.head() # This shows top rows. Default rows are 5.
df_1.head(n=3) # This shows top 3 rows
df_1.tail() # This shows bottom 5 rows
df_1.tail(n=4) # This shows bottom 4 rows
```

Many columns are omitted. If you want to see them, use `pd.set_option('display.max_columns', None)`

```python
pd.set_option('display.max_columns', None)
df_1.head(n=3)
```

If you want to check how a dataframe is big, use `df.shape`. It returns a set with data of (#rows, #columns).

```python
# Use comment and uncomment to execute each line
df_1.shape # This shows the size of df_1
df_2.shape # This shows the size of df_2
```

If you want to se  general statistical information, use `df.describe()`. It return a dataframe of count, mean, std and quantile values from numerical columns.

```python
print(df_1.shape)
df_1.describe()
```

If you use info() method, you can easily see how many rows are in the dataframe, the type of a each row, columns name, how many non-null data in a column.

```python
# Use comment and uncomment to execute each line
df_1.info()
df_2.info() 
```

There is time when you need to check the names of the columns or the indices in a dataframe. Use `df.columns` and `df.index`

```python
# Use comment and uncomment to execute each line
df_1.columns
df_2.index
```

It happens frequently you want to access a column or several columns to just see or modify them. There are 2 ways to do it. One is `df.column name`. The other is `df['column name']`. The first one is not able to be used when a column name has a empty space.

```python
# Use comment and uncomment to execute each line
# I used head() here to just to see some of rows.
print(df_1.columns[0:3])
df_1['Unique Key'].head()
df_1['Closed Date'].head()
df_1[['Closed Date','Unique Key']].head()
print(df_2.columns)
df_2.Rachel1.head()
```
